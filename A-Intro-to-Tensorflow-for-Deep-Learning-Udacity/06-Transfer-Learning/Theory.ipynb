## How do we perform Transfer Learning in practice

We cahnge the last layer of the neural network model. H
- The putput of the last layer must the Number of output class in our dataset
- Freezing the model - we do not want to change the pretrained part of the model during the training process. This is done by setting the variables of the pretrained model to non-trainable. So, we are making sure that only the variables from the last layer are trained. 





As an additional advantage, the training time is reduced significantly, as we are only training the variables in the last classification layer and not the entire model. 

# MobileNet

Some Neural Networks are very deep and contain millions of parameters which allow them to detect very complex patterns in the images. Large networks need a lot of training data, takes very long time to train, use a lot of memory and are very computationally expensive. 

State of the art Convolutional Neural Network, **MobileNet**, uses a very efficient Neural Network architecture which minimizes the amout of memory needed and computational resources needed while maintaining a very hgh level of accuracy. Ideal for mobile devices which have limited memory and computational resources. MobileNet was developed by Google and was trained on the imagenet dataset. with 1000 output classes. 

To perform transfer learning, we will download the features of the MobileNet without the classification layer. In tensorflow, the downloaded features can be used as regular KerasLayers with a particular input size. Since MobileNet was trained on ImageNet, we have to match the same input size that was used during training. (RGB 224*224)

- TensorFlow has a repository of pretrained models called TensorFlow hub.
- TensorFlow hub also has pretrained models where the last classification layer has been stripped out.

Here, you can see a version of MobileNet version two.

Using TensorFlow hub from code is really simple, you can just refer to the URL and then embed the pre-trained model as a KerasLayer in a sequential model. Then, we simply add our classification layer to the end of our model. Since our cat and dog dataset has only two classes, we'll add a dense layer with two outputs and a softmax activation function to our sequential model. This dense layer will be our new classifier and it needs to be trained.
